```python
import os
opb = os.path.basename
opd = os.path.dirname
from pathlib import Path
from datetime import datetime
from time import time
import random
import json
import shutil as sh
import argparse
##
from PIL import Image
##
import numpy as np
##
from matplotlib.backends.backend_agg import FigureCanvasAgg
from matplotlib.figure import Figure
from matplotlib import pyplot as plt
import cv2
```

```python
def save_gray(addr_img, output):
    '''
    Save in grayscale
    '''

    img = Image.open(addr_img).convert('L')
    img.save( Path(output) / opb(addr_img) ) 
                     
def read_extract_avi(film_addr):
    '''
    Read avi video and extract images
    '''
    
    vidcap = cv2.VideoCapture(film_addr)
    success, img = vidcap.read()
    count = 0
    try:
        os.mkdir('imgs')
    except:
        pass
    while success:
        addr_img = "imgs/frame{0}.{1}".format(count, 'tiff')
        print(addr_img)
        cv2.imwrite( addr_img, img )     # save frame in self.output_folder
        #save_gray(addr_img)         # triggered if self.args.gray_img True
        success, img = vidcap.read()
        count += 1
film_addr = '/home/meglio/Bureau/data/20190923/Position5-BF.avi'
read_extract_avi(film_addr)
```
